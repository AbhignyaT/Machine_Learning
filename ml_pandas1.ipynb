```python
import pandas as pd
import numpy as np
```

```python
#DF from list of lists
data=[['apple',10,240],['orange',15,150],['grapes',14,80],['pineapple',10,70]]
df1=pd.DataFrame(data,columns=['Fruit_name','Qty','Unit_price'])
df1
```

```python
#DF from dictionary
data={'Fruit_name':['apple','orange','grapes','pineapple'], 'Qty':[10,15,14,10], 'Unit_price':[240,150,80,70]}
df2=pd.DataFrame(data)
df2
```

```python
#DF from dictionary but with our own indices
data = {'Fruit_name':['apple', 'orange', 'grapes', 'pineapple'], 'Quantity':[10, 15, 14, 10], 'Unit_price':[240, 150, 80, 70]}
df3=pd.DataFrame(data,index=['f1','f2','f3','f4'])
df3
```

```python
#DF from list of dictionaries
data=[{'Fruit_name':'apple', 'Quantity':10,'Unit_price':240},{'Fruit_name':'orange', 'Quantity':15, 'Unit_price':150}, {'Fruit_name':'grapes', 'Quantity':14, 'Unit_price':80}, {'Fruit_name':'pineapple', 'Quantity':10, 'Unit_price':70}]
df4=pd.DataFrame(data)
df4
```

```python
#Adding a row, Method1
row={'Fruit_name':'litchy','Quantity':10, 'Unit_price':80}
df4=df4.append(row,ignore_index=True)     #ignore_index=True is must
df4
```

```python
#Adding row, Method2
df4.index

df4.loc[len(df4.index)]=['cherry',5,500]
df4
```

```python
#Adding multiple rows in dataframe
data=[{'Fruit_name':'mango', 'Quantity':13, 'Unit_price':140}, {'Fruit_name':'strawberry', 'Quantity':8, 'Unit_price':250}]
df5= pd.DataFrame(data)

df4=pd.concat([df4,df5],ignore_index=True)
df4
```

```python
#AddingColumn
df4['Total_price'] = df4['Quantity'] * df4['Unit_price']
df4
```

```python
#ChangingColumnNames
df4.columns=['Fruits','Qty','Uprice','Totprice'] #must have all the columns
df4
```

```python
#ChangingSingleColumnName
df4=df4.rename(columns={'Fruit':'Fruitname'})
df4
```

```python
#ChangingMultipleColumnNames
df4=df4.rename({'Qty':'Quantity','Uprice':'Unitprice'},axis='columns')
df4
```

```python
#Changing row names or index
df4.index=['f1','f2','f3','f8','f6','f4','f5','f7'] #must have all indices
df4
```

```python
#SortingRows based on row index
df4=df4.sort_index()
df4
```

```python
#Sorting rows descending order
df4=df4.sort_index(ascending=False) #descending=true doesn't work,wasnt defined
df4
```

```python
#Sorting cols on col names
df4=df4.sort_index(axis=1) #axis=0 sorts rows asce., index basis
df4
```

```python
#sorting cols, descending
df4=df4.sort_index(ascending=False,axis=1)
df4
# dff=df4.reset_index()
# dff
```

```python
#Creating series
s1=pd.Series(['apple', 'samsung', 'lenovo', 'dell', 'acer'],name='Brand')
s2=pd.Series([100000, 35000, 40000, 42000, 28000],name='price')
print(s1)
print(s2)
```

```python
#Creating DF from multiple series
df6=pd.merge(s2,s1,right_index=True,left_index=True) #the order of s1, s2,etc will be followed
df6
```

```python
#Converting single series to DF
df7=s1.to_frame()
df7
```

```python
#Single series to df with index as another col
df8=s1.to_frame().reset_index()
df8
```
